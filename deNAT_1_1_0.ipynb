<h2>Functions</h2>



```
def mapping_ip(ip_address):
  temp = ip_address.split('.')
  prefix = int(temp[0])
  if prefix < 128:
    return '.'.join(temp[:1])
  if prefix < 192:
    return '.'.join(temp[:2])
  if prefix < 224:
    return '.'.join(temp[:3])
  if prefix < 240:
    return 'multicasting'
  else:
    return 'experimentation'
def add_features(df):
  df['IP'] = df['IPV4_DST_ADDR'].fillna('').apply(mapping_ip)
  df['DURATION'] = df['FIRST_SWITCHED'] - df['LAST_SWITCHED']  
def encode(df_train, df_test):
  mapping = {}
  cat_features = ['IP']
  for feature in cat_features:
    temp = pd.Series(df_train[feature].unique())
    mapping[feature] = dict(zip(temp.values, temp.index))
    mapping[feature]['unknown'] = -1
    df_test[feature][~df_test[feature].isin(temp.values)] = 'unknown'
    df_train[feature] = df_train[feature].map(mapping[feature])
    df_test[feature] = df_test[feature].map(mapping[feature])
  return mapping
```

```
from abc import ABC, abstractmethod
# For all other model overwrite a evaluate method that return a data frame contains all results
class Model(ABC):
  def __init__(self):
    super().__init__()   
  @abstractmethod
  def evaluate(self) -> pd.DataFrame:
    pass
```

```
from sklearn.model_selection import train_test_split
import time
import lightgbm as lgb
import numpy as np
from sklearn.metrics import average_precision_score
from sklearn.metrics import roc_auc_score
class LGBM(Model):
  def __init(self):
    super().__init__()
  def evaluate(self, df_train, df_test, random_seed=1, num_round=50,
               params={'num_leaves':32, 'objective': 'binary', 'metric': 'auc'}):
    results = []
    for model in df_train['device_model'].unique():
      #print("//////////////////////////")
      #print(model+":")
      train = self.select(df_train, model)
      test = self.select(df_test, model)
      #print("{:.4%} in Training set".format(df_train[df_train['LABEL']==1].shape[0]/df_train.shape[0]))
      #print("{:.4%} in Testing set".format(df_test[df_test['LABEL']==1].shape[0]/df_test.shape[0]))
      percentage = (df_train[df_train['LABEL']==1].shape[0]/df_train.shape[0])

      x_train = train.drop(['LABEL'], axis=1).values
      y_train = train['LABEL'].values
      x_test = test.drop(['LABEL'], axis=1).values
      y_test = test['LABEL'].values
      # Ensure using the same val_datasets
      x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, 
                                  test_size=0.125, random_state=random_seed)
    
      #print(x_train.shape, x_val.shape, x_test.shape)
      train_data = lgb.Dataset(x_train, label=y_train, feature_name=['IN_PKTS', 'PROTOCOL', 'SRC_TOS', 'IN_BYTES', 'L4_SRC_PORT', 
              'L4_DST_PORT', 'TCP_FLAGS', 'IP', 'DURATION'], 
              categorical_feature = ['PROTOCOL', 'SRC_TOS', 'TCP_FLAGS', 'IP'])
      val_data = lgb.Dataset(x_val, y_val, reference=train_data)
      test_data = lgb.Dataset(x_test, y_test, reference=train_data)
      start_time = time.time()
      bst = lgb.train(params, train_data, num_round, valid_sets=[val_data], early_stopping_rounds=5)
      train_time = time.time() - start_time
      bst.save_model(model+".txt")
      val_scores = bst.predict(x_val, num_iteration=bst.best_iteration)
      val_AUPRC = average_precision_score(y_val, val_scores)
      val_ROC = roc_auc_score(y_val, val_scores)
      start_time = time.time()
      test_scores = bst.predict(x_test, num_iteration=bst.best_iteration)
      prediction_time = time.time() - start_time
      test_AUPRC = average_precision_score(y_test, test_scores)
      test_ROC = roc_auc_score(y_test, test_scores)
      results.append([model, percentage, val_AUPRC, val_ROC, test_AUPRC, test_ROC, train_time, prediction_time])
    results = pd.DataFrame(results, columns=['Model', 'Percentage', 'Val AUPRC', 'Val ROC', 
                                           'Test AUPRC', 'Test ROC', 'Training Time', 'Prediction Time'])
    return results
  
  def select(slef, df, model):
    df['LABEL'] = (df['device_model']==model).astype(int) 
    return df[['IN_PKTS', 'PROTOCOL', 'SRC_TOS', 'IN_BYTES', 'L4_SRC_PORT', 
             'L4_DST_PORT', 'TCP_FLAGS', 'IP', 'DURATION', 'LABEL']]
```

```
# num_h1, num_h2, epochs
# Support validation early stopping callback on validation set
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.optimizers import RMSprop
from sklearn.preprocessing import MinMaxScaler
from keras.callbacks import EarlyStopping
class DNN(Model):
  def __init__(self):
    super().__init__()
  def evaluate(self, df_train, df_test, random_seed, num_h1=6, num_h2=6, epochs=15):
    results = []
    for model in df_train['device_model'].unique():
      train = self.select(df_train, model)
      test = self.select(df_test, model)
      percentage = (df_train[df_train['LABEL']==1].shape[0]/df_train.shape[0])
      x_train = train.drop(['LABEL'], axis=1).values
      y_train = train['LABEL'].values
      x_test = test.drop(['LABEL'], axis=1).values
      y_test = test['LABEL'].values
      x_train, x_val, y_train, y_val = train_test_split(x_train, y_train, 
                                  test_size=0.125, random_state=random_seed)
      x_train, x_val, x_test = self.normalize(x_train, x_val, x_test)
      dnn = Sequential()
      dnn.add(Dense(num_h1, activation='relu', input_shape=(9, )))
      dnn.add(Dropout(0.1))
      dnn.add(Dense(num_h2, activation='relu'))
      dnn.add(Dropout(0.1))
      dnn.add(Dense(1, activation='sigmoid'))
      dnn.compile(loss='binary_crossentropy', optimizer=RMSprop(), metrics=['accuracy'])

      es = EarlyStopping(monitor='val_loss')
      start_time = time.time()
      dnn.fit(x_train, y_train, validation_data=(x_val, y_val), epochs=epochs, batch_size=256,
                callbacks=[es])
      train_time = time.time() - start_time
      dnn.save_weights(model+'.h5')
      val_scores = dnn.predict(x_val)
      val_AUPRC = average_precision_score(y_val, val_scores)
      val_ROC = roc_auc_score(y_val, val_scores)
      start_time = time.time()
      test_scores = dnn.predict(x_test)
      prediction_time = time.time() - start_time
      test_AUPRC = average_precision_score(y_test, test_scores)
      test_ROC = roc_auc_score(y_test, test_scores)
      results.append([model, percentage, val_AUPRC, val_ROC, test_AUPRC, test_ROC, train_time, prediction_time])
    results = pd.DataFrame(results, columns=['Model', 'Percentage', 'Val AUPRC', 'Val ROC', 'Test AUPRC', 'Test ROC', 'Training Time', 'Prediction Time'])
    return results

  def normalize(self, x_train, x_val, x_test):
    scaler = MinMaxScaler()
    x_train = scaler.fit_transform(x_train)
    x_val = scaler.transform(x_val)
    x_test = scaler.transform(x_test)
    return x_train, x_val, x_test

  def select(slef, df, model):
    df['LABEL'] = (df['device_model']==model).astype(int) 
    return df[['IN_PKTS', 'PROTOCOL', 'SRC_TOS', 'IN_BYTES', 'L4_SRC_PORT', 
             'L4_DST_PORT', 'TCP_FLAGS', 'IP', 'DURATION', 'LABEL']]
```

```
# SVM Do not support validation sets
# SVM using LinearSVR
# Rationale Behind not using RBF Kernel： Too Slow, O(n^3)

from joblib import dump
from sklearn import svm
class SVM(Model):
  def __init__(self):
    super().__init__()
  def evaluate(self, df_train, df_test, random_state, C=1, kernel='rbf'):
    results = []
    for model in df_train['device_model'].unique():
      train = self.select(df_train, model)
      test = self.select(df_test, model)
      percentage = (df_train[df_train['LABEL']==1].shape[0]/df_train.shape[0])
      x_train = train.drop(['LABEL'], axis=1).values
      y_train = train['LABEL'].values
      x_test = test.drop(['LABEL'], axis=1).values
      y_test = test['LABEL'].values
      x_train, x_test = self.normalize(x_train, x_test)
      clf = svm.LinearSVC(C=C, random_state=random_state)
      start_time = time.time()
      clf.fit(x_train, y_train)
      train_time = time.time() - start_time
      dump(clf, "{}_svm.joblib".format(model))
      start_time = time.time()
      test_scores = clf.decision_function(x_test)
      prediction_time = time.time() - start_time
      test_AUPRC = average_precision_score(y_test, test_scores)
      test_ROC = roc_auc_score(y_test, test_scores)
      results.append([model, percentage, test_AUPRC, test_ROC, train_time, prediction_time])
    results = pd.DataFrame(results, columns=['Model', 'Percentage','Test AUPRC', 'Test ROC', 'Training Time', 'Prediction Time'])
    return results

  def normalize(sellf, x_train, x_test):
    scaler = MinMaxScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)
    return x_train, x_test

  def select(slef, df, model):
    df['LABEL'] = (df['device_model']==model).astype(int) 
    return df[['IN_PKTS', 'PROTOCOL', 'SRC_TOS', 'IN_BYTES', 'L4_SRC_PORT', 
             'L4_DST_PORT', 'TCP_FLAGS', 'IP', 'DURATION', 'LABEL']]
```

```
# Do not support validation
from sklearn.ensemble import IsolationForest
class IF(Model):
  def __init__(self):
    super().__init__()
  
  def evaluate(self, df_train, df_test, random_state):
    results = []
    for model in df_train['device_model'].unique():
      train = self.select(df_train, model)
      test = self.select(df_test, model)
      percentage = (df_train[df_train['LABEL']==1].shape[0]/df_train.shape[0])
      x_train = train.drop(['LABEL'], axis=1).values
      y_train = train['LABEL'].values
      x_test = test.drop(['LABEL'], axis=1).values
      y_test = test['LABEL'].values
      clf = IsolationForest(random_state=random_state, n_estimators=100)
      start_time = time.time()
      clf.fit(x_train)
      train_time = time.time() - start_time
      dump(clf, "{}_if.joblib".format(model))
      start_time = time.time()
      test_scores = clf.decision_function(x_test)
      prediction_time = time.time() - start_time
      test_AUPRC = average_precision_score(y_test, test_scores)
      test_ROC = roc_auc_score(y_test, test_scores)
      results.append([model, percentage, test_AUPRC, test_ROC, train_time, prediction_time])
    results = pd.DataFrame(results, columns=['Model', 'Percentage','Test AUPRC', 'Test ROC', 'Training Time', 'Prediction Time'])
    return results  

  def select(slef, df, model):
    df['LABEL'] = (df['device_model']==model).astype(int) 
    return df[['IN_PKTS', 'PROTOCOL', 'SRC_TOS', 'IN_BYTES', 'L4_SRC_PORT', 
             'L4_DST_PORT', 'TCP_FLAGS', 'IP', 'DURATION', 'LABEL']]
```

```
# Auto Encoder Do not support valiadatraion
class AutoEncoder(Model):
  def __init__(self, layout, p_drop):
    super().__init__()
    self.layout = layout
    self.p_drop = p_drop
  
  def get_model(self):
    model = Sequential()
    if(self.layout[0]!=self.layout[-1]):
      raise Exception("input dimension should match the output dimension")
    model.add(Dense(self.layout[1], input_shape=(self.layout[0], )))
    model.add(Dropout(self.p_drop))
    for dim in self.layout[2:-1]:
      model.add(Dense(dim, activation='relu'))
      model.add(Dropout(self.p_drop))
    model.add(Dense(self.layout[-1], activation='sigmoid'))
    model.compile(optimizer='adam', loss='mean_squared_error')
    return model

  def evaluate(self, df_train, df_test):
    results = []
    for model in df_train['device_model'].unique():
      train = self.select(df_train, model)
      test = self.select(df_test, model)
      percentage = (df_train[df_train['LABEL']==1].shape[0]/df_train.shape[0])
      x_train = train.drop(['LABEL'], axis=1).values
      y_train = train['LABEL'].values
      x_test = test.drop(['LABEL'], axis=1).values
      y_test = test['LABEL'].values
      x_train, x_test = self.normalize(x_train, x_test)
      x_0 = x_train[y_train==0, :]
      start_time = time.time()
      ae = self.get_model()
      ae.fit(x_0, x_0, epochs=5, batch_size=256,
                    shuffle=True,
                    verbose=1)
      train_time = time.time() - start_time
      ae.save_weights(model+'_ae.h5')
      start_time = time.time()
      test_predict = ae.predict(x_test)
      test_scores = np.square(x_test - test_predict).mean(axis = 1)
      prediction_time = time.time() - start_time
      test_AUPRC = average_precision_score(y_test, test_scores)
      test_ROC = roc_auc_score(y_test, test_scores)
      results.append([model, percentage, test_AUPRC, test_ROC, train_time, prediction_time])
    results = pd.DataFrame(results, columns=['Model', 'Percentage','Test AUPRC', 'Test ROC', 'Training Time', 'Prediction Time'])
    return results

  def normalize(sellf, x_train, x_test):
    scaler = MinMaxScaler()
    x_train = scaler.fit_transform(x_train)
    x_test = scaler.transform(x_test)
    return x_train, x_test

  def select(slef, df, model):
    df['LABEL'] = (df['device_model']==model).astype(int) 
    return df[['IN_PKTS', 'PROTOCOL', 'SRC_TOS', 'IN_BYTES', 'L4_SRC_PORT', 
             'L4_DST_PORT', 'TCP_FLAGS', 'IP', 'DURATION', 'LABEL']]
```
